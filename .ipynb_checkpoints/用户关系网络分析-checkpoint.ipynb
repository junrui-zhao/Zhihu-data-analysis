## 2.最小二乘法拟合指数截断的幂律拟合，双对数坐标展示

```python
def powerLawExponentialCutOffPlot(datax,datay, xlab, ylab,left,right):
    '''
    Plot fitted powerLaw distribution with Exponential CutOff
    '''
    t = datax
    #r = np.array(range(len(data))) +1
    r = datay
    y = np.log(r[left:right])
    x1 = np.log(t[left:right])
    x2 = t[left:right]
    x = np.column_stack((x1,x2))
    x = sm.add_constant(x, prepend=True)
    res = sm.OLS(y,x).fit()
    L,alpha,lambde = res.params
    r2 = res.rsquared
    plt.plot(t,r,".",color="blue",alpha=1)
    plt.plot(t[left:right], (np.exp(L) * t ** alpha * np.exp(lambde * t))[left:right],"r-",linewidth=2)
    plt.xscale('log'); plt.yscale('log')
    plt.ylim(ymax = 0.043)
    plt.xlabel(xlab, fontsize = 10)
    plt.ylabel(ylab, fontsize = 10)
    plt.text(3, 0.001, "alpha="+ str(round(alpha,3))+"\nlambda="+str(round(lambde,3)) ,fontsize=10);
    plt.show()
    print L,alpha,lambde, r2
```

### 3.3 双对数坐标下粉丝数概率统计（回归拟合幂律分布）

```python
def drawpic(X,pY):
    _logX = []
    _logpY = []

    for x, y in zip(X,pY): #这里可以调整
        _logX.append([float(np.log10(x))])
        _logpY.append([float(np.log10(y))])
    # 模型拟合
    regr = linear_model.LinearRegression()
    regr.fit(_logX[4000:len(X)-10], _logpY[4000:len(pY)-10])
    # 模型结果与得
    print('Coefficients:', regr.coef_,)
    print("Intercept:",regr.intercept_)
# The mean square error
    print("Residual sum of squares: %.8f"
      % np.mean((regr.predict(_logX) - _logpY) ** 2))  # 残差平方和

    # 可视化
    #plt.title("following count loglog probability distribution")
    plt.xlabel("the number of followers")
    plt.ylabel("p(k)")
    
    plt.plot(X,pY,".",  color='blue')
    plt.plot(X[4000:len(X)-15], X[4000:len(X)-15]**regr.coef_[0][0]*np.exp(regr.intercept_[0]), "r-",linewidth=2)
    #print plt.yticks()
    #plt.yticks(plt.yticks()[0],['$10^%f$'%w for w in range(3)]) 
    #plt.yticks()[0]
    plt.xscale('log'); plt.yscale('log')
    plt.text(5, 0.001, "Slope="+ str(round(regr.coef_[0][0],2)) ,fontsize=10);
    # plt.xticks(())
    # plt.yticks(())
    plt.show()
drawpic(np.array(X),np.array(pY))
```

### 3.4 用户关系网络

```python
G = nx.DiGraph()
Location = r'../数据/论文用/普通用户关注关系id-id.csv'
#df = pd.read_csv(Location,dtype={'user_detail.userid':np.str,'user_detail.following_count':np.int32})
df = pd.read_csv(Location,dtype={'Source':np.str,'Target':np.str})
user_relation = []
user_id = df['Target']
userid_follower = df['Source']
user_relation = zip(list(userid_follower),list(user_id))
```

```python
for row in user_relation:
    G.add_edges_from([(row[0],row[1])])
```

```python
values = G.in_degree().values()
value_count_pairs = collections.Counter(values).most_common()
sorted_pairs = sorted(value_count_pairs)[1:]
X  = zip(*sorted_pairs)[0]
# Normalize the data to a proper PDF
Y = zip(*sorted_pairs)[1]
pY = [float(float(y)/float(sum(Y))) for y in zip(*sorted_pairs)[1]]
X,pY
```

### 3.4.1 入度分布图

```python
#画图+回归拟合（特征，概率）
def drawpic(l,r,X,pY,x,y):
    logX = []
    logpY = []
    _logX = []
    _logpY = []

    for x, y in zip(X,pY):
        logX.append([float(np.log10(x))])
        logpY.append([float(np.log10(y))])

    for x, y in zip(X[l:r],pY[l:r]): #这里可以调整
        _logX.append([float(np.log10(x))])
        _logpY.append([float(np.log10(y))])
    # 模型拟合
    regr = linear_model.LinearRegression()
    regr.fit(_logX, _logpY)
    # 模型结果与得分
    print('Coefficients:', regr.coef_,)
    print("Intercept:",regr.intercept_)
    # The mean square error
    print("Residual sum of squares: %.8f"
      % np.mean((regr.predict(_logX) - _logpY) ** 2))  # 残差平方和

    # 可视化
    #plt.title("following count loglog probability distribution")
    plt.xlabel("in degree")
    plt.ylabel("P(k)")
    plt.plot(logX,logpY,".",  color='blue')
    plt.plot(_logX, regr.predict(_logX), color='red',linewidth=2)
    plt.text(1.8, -4, "Slope="+ str(round(regr.coef_[0][0],2)) ,fontsize=10);
    # plt.xticks(())
    # plt.yticks(())
    plt.show()
drawpic(40,120,X,pY,1.8,-4)
```
